## How To Work With Delimeted Format?

```python
!type data\dqtext.csv 
```

```python
import csv 
csv.reader?
```

```python
f = open('data/dqtext.csv')
reader = csv.reader(f)
reader
```

```python
for line in reader: 
    print(line)
```

```python
with open('data/dqtext.csv') as f: 
    lines = list(csv.reader(f)) 
    
print(lines)
```

```python
header, values = lines[0], lines[1:]
```

```python
data_dict = {i: v for i, v in zip(header, zip(*values))}
data_dict
```

```python
header, values = lines[0], lines[1:]
print(header, values, end='\n')
```

```python
csv.writer?
```

```python
with open('data/myfile2.csv', 'w') as f: 
    writer = csv.writer(f) 
    writer.writerow(('one', 'two', 'three')) 
    writer.writerow(('1', '2', '3')) 
    writer.writerow(('4', '5', '6')) 
    writer.writerow(('7', '8', '9'))
```
